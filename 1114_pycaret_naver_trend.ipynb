```python
import pandas as pd
from pycaret.classification import *
from pycaret.datasets import get_data
data = get_data('wine')
```

```python
data.type.value_counts()
```

```python
df = pd.read_csv('naver_mean_0_1_history_labeled.csv')
df.head()
```

```python
df['target'] = df.target.astype('str')
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split( df.iloc[:, 2:], test_size=0.2, stratify = df.target, random_state=2022,)
train.shape, test.shape
```

```python
clf = setup(data = train, target = 'target', silent=True)
```

```python
top5_models = compare_models(fold=5, round=3, sort='f1', n_select=5)
```

```python
blender = blend_models(top5_models, choose_better=True)
```

```python
tuned_blender = tune_model(blender)
```

```python
predictions = predict_model(tuned_blender, data=test)
predictions.head()
```
