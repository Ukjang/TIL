```python
import pandas as pd
from pycaret.classification import *
from pycaret.datasets import get_data
data = get_data('wine')
```

```python
data.type.value_counts()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split( data, test_size=0.2, stratify = data.type, random_state=2022,)
train.shape, test.shape
```

```python
clf = setup(data = train, target = 'type', silent=True, fix_imbalance=True)
```

```python
top5_models = compare_models(fold=5, round=3, sort='f1', n_select=5)
```

```python
blender = blend_models(top5_models, choose_better=True)
```

```python
tuned_blender = tune_model(blender)
```

```python
predictions = predict_model(tuned_blender, data=test)
predictions.head()
```
